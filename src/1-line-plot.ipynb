### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values ranging from 6 to 10 (inclusive)
x = np.arange(6, 11)  # 11 is exclusive, so it goes up to 10

# Calculate y values (y = x^2)
y = x**2

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title("Line Plot of y = x^2 (x: 6 to 10)") # Set the title
plt.xlabel("x-axis")  # Label the x-axis
plt.ylabel("y-axis")  # Label the y-axis
plt.grid(True)  # Turn on the grid

# Display the plot
plt.show()
```
